```python
c_type=[]
c_list=[]
c_link=[]
c_code=[]
work=[]

for i in range (1,25):
    website_url_all = 'https://catalog.belmont.edu/content.php?catoid=9&catoid=9&navoid=445&filter%5Bitem_type%5D=3&filter%5Bonly_active%5D=1&filter%5B3%5D=1&filter%5Bcpage%5D=' +str(i)
    print(website_url_all)
    response_all = requests.get(website_url_all)
    print(response_all.status_code)
    response_all.content
    soup_all = BeautifulSoup(response_all.content, 'lxml')
    for type in soup_all.find_all('p'):
        print(type.string)
        c_type.append(type.string)
    for link in soup_all.find_all('td', {'class': 'width'}, 'href'):
        print("{}".format(link.text[6:-1]))
        print(link.a.get("href"))
        print(link.a.get("href")[-5:])
        c_list.append("{}".format(link.text[6:-1]))
        c_link.append("{}".format(link.a.get("href")))
        c_code.append("{}".format(link.a.get("href")[-5:]))
```
