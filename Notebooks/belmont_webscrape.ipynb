```python
belmont_main = 'https://www.belmont.edu/'
response_main = requests.get(belmont_main)
response_main.status_code
```

```python
print(type(response_main))
response_main.content
```

```python
soup_main = BeautifulSoup(response_main.content, 'lxml')
print(soup_main.title)
```
