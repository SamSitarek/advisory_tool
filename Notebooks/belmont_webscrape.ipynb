# Belmont Degrees & Programs

```python
belmont_programs = 'https://catalog.belmont.edu/content.php?catoid=9&navoid=444'
response_programs = requests.get(belmont_programs)
response_programs.status_code
```

```python
print(type(response_programs))
response_programs.content
```

```python
soup_programs = BeautifulSoup(response_programs.content, 'lxml')
print(soup_programs.title)
```

```python
print(soup_programs)
```

```python
print(soup_programs.prettify())
```

```python
degree_type=[]


for type in soup_programs.find_all('p', attrs = {'style': 'padding-left: 30px'}):
        print(type.string)
        degree_type.append(type.string)
```

```python
print(degree_type)
```

```python
family=[]

for type in soup_programs.find_all('li', attrs = {'style': 'list-style-type: none'}):
    print(list(type.children))
    family.append(list(type.children))
```

```python
len(family)
```

```python
maj=[]
maj_link=[]
 
for link in soup_programs.find_all('li', attrs = {'style': 'list-style-type: none'}):
    print("{}".format(link.text))
    print(link)
    maj.append("{}".format(link.text))
    maj_link.append(link)
```

```python
maj
```

```python
maj_link
```

```python
maj_links=[]

for each in maj_link:
    print(str(each.a))
    maj_links.append(str(each.a))
```

```python
splitted_maj_links=[]

for each in maj_links:
    print(each.split('>'))
    splitted_maj_links.append(each.split('>'))
```

```python
len(splitted_maj_links)
```

```python
splitted_maj_links[209][0]
```

```python
front=[]
back=[]

for i in range(0,210):
    f= splitted_maj_links[i][0]
    b= splitted_maj_links[i][1]
    print(f)
    print(b)
    front.append(f)
    back.append(b)
    print(len(front))
    print(len(back))
```

```python
program=[]

for each in front:
    print(each.split('"'))
    program.append(each.split('"')[1])
```

```python
program
```

```python
program_link=[]

for i in range(0,210):
    p=program[i]
    print(str('https://catalog.belmont.edu/')+p)
    program_link.append(str('https://catalog.belmont.edu/')+p)
    print(len(program_link))
```

```python
program_link
```

```python
major=[]

for each in back:
    print(each.split('<'))
    major.append(each.split('<')[0])
```

```python
belmont=pd.DataFrame(columns=['Degree', 'Program'])
```

```python
counter = 0
counter2 = 0
hold = "aa"
architect=[0]
n_s=[108,112]
minor=[149,199]

for row in range(0,210):
    a=major[row]
    if row in architect:
        belmont.loc[counter, 'Degree']=degree_type[counter2]
        belmont.loc[counter,'Program']= a
        counter+=1
        counter2+=1
        hold = a
    elif row in n_s:
        counter2+=1
        belmont.loc[counter, 'Degree']=degree_type[counter2]
        belmont.loc[counter,'Program']= a
        counter+=1
        hold = a
    elif row in minor:
        belmont.loc[counter, 'Degree']=degree_type[counter2]
        belmont.loc[counter,'Program']= a
        counter+=1
        hold = a
    elif hold < a:
        belmont.loc[counter, 'Degree']= degree_type[counter2]
        belmont.loc[counter, 'Program'] = a
        counter += 1
        hold = a
    else:
        counter2 +=1
        belmont.loc[counter, 'Degree']= degree_type[counter2]
        belmont.loc[counter, 'Program'] = a
        counter += 1
        hold = a
```
